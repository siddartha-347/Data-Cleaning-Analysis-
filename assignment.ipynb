```python
import pandas as pd
duplicated_movies = movies_df[movies_df.duplicated('MOVIES', keep=False)]

def handle_duplicates(group):
    if group.empty:  
        return pd.Series(dtype=object)
    original = group.iloc[0].copy()

    original['RunTime'] = group['RunTime'].max()
 
    total_votes = group['VOTES'].sum()
    original['VOTES'] = total_votes

    if total_votes > 0:
        original['RATING'] = (group['RATING'] * group['VOTES']).sum() / total_votes
    else:
        original['RATING'] = group['RATING'].mean()


    for col in ['STARS', 'ONE-LINE', 'GENRE', 'YEAR']:
        valid_index = group[col].first_valid_index()
        if valid_index is not None:
            original[col] = group.loc[valid_index, col]
        else:
            original[col] = None  

    return original


cleaned_movies = (
    duplicated_movies.groupby('MOVIES', group_keys=False)
    .apply(handle_duplicates)
    .reset_index(drop=True)  
)

movies_df = movies_df.drop(duplicated_movies.index)

movies_df = pd.concat([movies_df, cleaned_movies], ignore_index=True)


print(movies_df.info())
print(movies_df.describe())
print(movies_df.head())
```

```python
import re
import pandas as pd
import numpy as np

def split_year(year):
    current_year = 2025
    
    if pd.isnull(year):
        return pd.Series([np.nan, np.nan])
    
    year_str = str(year).strip()
    
    if '-' in year_str or '–' in year_str:
        parts = re.split(r'[-–]', year_str, 1)
        from_part = parts[0].strip(' ()')
        to_part = parts[1].strip(' ()') if len(parts) > 1 else ''
        
        from_year = re.search(r'\d{4}', from_part)
        to_year = re.search(r'\d{4}', to_part)
        
        return pd.Series([
            int(from_year.group()) if from_year else np.nan,
            int(to_year.group()) if to_year else current_year
        ])
    
    year_match = re.search(r'\d{4}', year_str)
    if year_match:
        year_value = int(year_match.group())
        return pd.Series([year_value, year_value])
    
    return pd.Series([np.nan, np.nan])

movies_df[['From Year', 'To Year']] = movies_df['YEAR'].apply(split_year)
movies_df[['From Year', 'To Year']] = movies_df[['From Year', 'To Year']].astype('Int64')
```

```python
avg_runtime_from_year = movies_df.groupby('From Year')['RunTime'].mean()
avg_runtime_to_year = movies_df.groupby('To Year')['RunTime'].mean()

def fill_years_from_runtime(row):
    if pd.notnull(row['From Year']) and pd.notnull(row['To Year']):
        return row['From Year'], row['To Year']
    
    if pd.notnull(row['RunTime']):
        closest_from_year = avg_runtime_from_year.sub(row['RunTime']).abs().idxmin()
        closest_to_year = avg_runtime_to_year.sub(row['RunTime']).abs().idxmin()
        
        return closest_from_year, closest_to_year
    overall_avg_runtime = (avg_runtime_from_year.mean() + avg_runtime_to_year.mean()) / 2
    closest_from_year = avg_runtime_from_year.sub(overall_avg_runtime).abs().idxmin()
    closest_to_year = avg_runtime_to_year.sub(overall_avg_runtime).abs().idxmin()
    
    return closest_from_year, closest_to_year

movies_df[['From Year', 'To Year']] = movies_df.apply(lambda row: fill_years_from_runtime(row) if pd.isnull(row['From Year']) or pd.isnull(row['To Year']) else (row['From Year'], row['To Year']), axis=1, result_type='expand')

movies_df[['From Year', 'To Year']] = movies_df[['From Year', 'To Year']].astype('Int64')

movies_df.head()
```

```python
avg_runtime_from_year = movies_df.groupby('From Year')['RunTime'].mean()
avg_runtime_to_year = movies_df.groupby('To Year')['RunTime'].mean()

def fill_runtime(row):
    if pd.notnull(row['RunTime']):
        return row['RunTime']
    
    from_year_runtime = avg_runtime_from_year.get(row['From Year'], np.nan)
    to_year_runtime = avg_runtime_to_year.get(row['To Year'], np.nan)
    
    if pd.notnull(from_year_runtime) and pd.notnull(to_year_runtime):
        return (from_year_runtime + to_year_runtime) / 2
    elif pd.notnull(from_year_runtime):
        return from_year_runtime
    elif pd.notnull(to_year_runtime):
        return to_year_runtime
    else:
        return np.nan

movies_df['RunTime'] = movies_df.apply(fill_runtime, axis=1)
movies_df.head()
```

```python
def count_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = series[(series < lower_bound) | (series > upper_bound)]
    return outliers.count()

rating_outliers = count_outliers(movies_df['RATING'])
votes_outliers = count_outliers(movies_df['VOTES'])
runtime_outliers = count_outliers(movies_df['RunTime'])

print(f"Number of outliers in RATING: {rating_outliers}")
print(f"Number of outliers in VOTES: {votes_outliers}")
print(f"Number of outliers in RunTime: {runtime_outliers}")
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
movies_df.boxplot(column='RATING', ax=axes[0])
axes[0].set_title('Box Plot of RATING')
movies_df.boxplot(column='VOTES', ax=axes[1])
axes[1].set_title('Box Plot of VOTES')
movies_df.boxplot(column='RunTime', ax=axes[2])
axes[2].set_title('Box Plot of RunTime')

plt.tight_layout()
plt.show()
```

```python
def handle_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    near_lower_bound = Q1 - 3 * IQR
    near_upper_bound = Q3 + 3 * IQR

    def approximate_value(value):
        if value < lower_bound:
            if value >= near_lower_bound:
                return lower_bound
            else:
                return np.nan
        elif value > upper_bound:
            if value <= near_upper_bound:
                return upper_bound
            else:
                return np.nan
        else:
            return value

    return series.apply(approximate_value)

movies_df['RATING'] = handle_outliers(movies_df['RATING'])
movies_df['VOTES'] = handle_outliers(movies_df['VOTES'])
movies_df['RunTime'] = handle_outliers(movies_df['RunTime'])

movies_df.dropna(subset=['RATING', 'VOTES', 'RunTime'], inplace=True)

movies_df.reset_index(drop=True, inplace=True)
movies_df.info()
movies_df.head()
```
